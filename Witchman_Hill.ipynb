```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math  
```

```
n = 1000 #Number of random numbers to generate 100,1000,5000
```

#LCG-1

```
u1 = []
z1 = [12, 7]
m1 = 16
```

```
for i in range(2,n+2):
  temp = (13*z1[i-1] + 11*z1[i-2] + 3)%m1
  z1.append(temp)
  u1.append(z1[i]/m1)
print(len(z1))
print(len(u1))
```

#LCG-2

```
u2 = []
z2 = [3, 5]
m2 = 17
```

```
for i in range(2,n+2):
  temp = (12*z2[i-1]**2 + 13*z2[i-2])%m2
  z2.append(temp)
  u2.append(z2[i]/m2)
print(len(z2))
print(len(u2))
```

#LCG-3

```
u3 = []
z3 = [2, 7]
m3 = 15
```

```
for i in range(2,n+2):
  temp = (z3[i-1]**3 + z3[i-2]**2)%m3
  z3.append(temp)
  u3.append(z3[i]/m3)
print(len(z3))
print(len(u3))
```

#Combining  G1, G2 and G3 together to generate random numbers


```
u = []
for i in range(0,n):
  temp_u = u1[i]+u2[i]+u3[i]
  u.append(temp_u - int(temp_u))
print(u)
```

#Visualization

```
plt.figure(figsize=(50,15))
indices = range(0,len(u))
plt.bar(indices,u)
plt.xlabel('indices') 
plt.ylabel('random numbers') 
plt.title('Witchman / Hill') 
plt.show() 
```
